```python
import numpy as np
import pandas as pd

df = pd.read_csv("data.csv", encoding="latin1")

df = df.dropna(subset=["no2"])
x = df["no2"].values.astype(float)

print(f"Total samples used: {len(x)}")
```

```python
r = 102316089

ar = 0.05 * (r % 7)
br = 0.3 * (r % 5 + 1)

print(f"ar = {ar}")
print(f"br = {br}")
```

```python
z = x + ar * np.sin(br * x)

mu = np.mean(z)
vari = np.mean((z - mu) ** 2)

lamda = 1 / (2 * vari)
c = np.sqrt(lamda / np.pi)

print(f"μ (mean) = {mu}")
print(f"λ (lambda) = {lamda}")
print(f"c (constant) = {c}")
```
